```python
import selenium
from selenium.webdriver import Chrome
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import time

driver = Chrome('./chromedriver')
url = 'https://www.zoopla.co.uk/to-rent/property/liverpool/?beds_min=1&page_size=25&price_frequency=per_month&price_max=700&view_type=list&q=Liverpool%2C%20Merseyside&radius=0&results_sort=newest_listings&search_source=refine'
# Scraper Class to Navigate Website
driver.get(url)
time.sleep(3)
full_prop_list = []
pages = 0

def find_and_accept_cookies():
    try:
        driver.switch_to.frame('gdpr-consent-notice') # This is the id of the frame
        accept_cookies_button = driver.find_element_by_xpath('//*[@id="save"]')
        accept_cookies_button.click()
        time.sleep(1)
    except AttributeError: # If you have the latest version of selenium, the code above won't run because the "switch_to_frame" is deprecated
        driver.switch_to_frame('gdpr-consent-notice') # This is the id of the frame
        accept_cookies_button = driver.find_element_by_xpath('//*[@id="save"]')
        accept_cookies_button.click()
        time.sleep(1)

    except:
        pass
    return driver


time.sleep(10)

prop_container = driver.find_element_by_xpath('//div[@class="css-1anhqz4-ListingsContainer e1b8efd72"]')
prop_list = prop_container.find_elements_by_xpath('./div')
num_prop = len(prop_list)

for i in range(num_prop):
        prop_container = driver.find_element_by_xpath('//div[@class="css-1anhqz4-ListingsContainer e1b8efd72"]')
        house = prop_container.find_elements_by_xpath('./div')[i]
        data = {i: {"rent_price": [], "weekly_price": [],"bed_and_bath_num": [], "description": [], "address": []}}
        house.click()
        time.sleep(10) # To make sure the webpage has time to load
            
          # Try to find each corresponding information and append 'None' if not present
        try:
            rent_price = driver.find_element_by_xpath('//p[@data-testid = "price"]').text
            data[i]['rent_price'].append(rent_price)
        except NoSuchElementException:
            data[i]['rent_price'].append('None')

        try:
            weekly_price = driver.find_element_by_xpath('//p[@data-testid="rentalfrequency-and-floorareaunit"]').text
            data[i]['weekly_price'].append(weekly_price)
        except NoSuchElementException:
            data[i]['weekly_price'].append('None')

        try:
            bed_and_bath_num = driver.find_element_by_xpath('//div[@class="c-PJLV c-PJLV-iiNveLf-css"]').text
            data[i]['bed_and_bath_num'].append(bed_and_bath_num)
        except NoSuchElementException:
            data[i]['bed_and_bath_num'].append('None')

                
        try:
            description = driver.find_element_by_xpath('//div[@data-testid="listing_description"]').text
            data[i]['description'].append(description)
        except NoSuchElementException:
            data[i]['description'].append('None')

        try:
            address = driver.find_element_by_xpath('//address[@ data-testid="address-label"]').text
            data[i]['address'].append(address)
        except NoSuchElementException:
            data[i]['address'].append('None')
                
        full_prop_list.append(data)
        print(full_prop_list[i])
        time.sleep(5)
        driver.back()
print(full_prop_list)
next_page_buttom = driver.find_element_by_xpath('//a[@class="eaoxhri5 css-xtzp5a-ButtonLink-Button-StyledPaginationLink eaqu47p1"]')
       

```
